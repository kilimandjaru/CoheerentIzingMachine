```python
from cim_lib import discrete_CIM as CIM
from cim_lib import plot_spin_beh as plot
from cim_lib import dwave, parser

import numpy as np
import matplotlib.pyplot as plt
```

```python
cim = CIM(alpha1 = 0.9,
            alpha2 = 0.90,
            b = -0.7,
            J = parser("dataset\G12.txt"))
```

```python
iters = 20
state, energy = cim.anneal(iters, seed = 0)
print(min(energy))
plt.scatter(range(iters), energy)
```

```python
d_vec = dwave(matrix = cim.J, num_reads=1000)
d_vec.dot(parser("dataset\G12.txt")).dot(d_vec.T)
```
